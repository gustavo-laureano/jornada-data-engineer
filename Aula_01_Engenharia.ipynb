### Bloco 1 - Simulando a camada Bronze/Raw

```python
from pyspark.sql.types import *
# Criando dados na memória (simulando um arquivo que chegou)
data = [
(1, "Gustavo", "TI", 2500),
(2, "Alexandre", "TI", 2500),
(3, "Chico", "Engenharia", 3000),
(4, "Chefe", "Gestão", 15000),
(5, "Estagiario", "TI", None) # Dado sujo/nulo
]
columns = ["id", "nome", "departamento", "salario"]
# Criando o DataFrame A tabela distribuída do Spark)
df_bronze = spark.createDataFrame(data, columns)
# Visualizando
display(df_bronze)
```

### Tarefa: Alterar o código para adicionar mais 2 pessoas na lista inicial.

```python
data.extend([
    (6, "Nicolas", "TI", 5000),
    (8, "Neymar", "Joga de bola", 1000000)
])


columns = ["id", "nome", "departamento", "salario"]
df_bronze = spark.createDataFrame(data, columns)

display(df_bronze)
```

### Bloco 2 - Transformação com PySpark SQL Simulando a camada Silver
_+ alteração tarefa._

```python
from pyspark.sql.functions import col

#Alteração task > Quem for do departamento "Gestão" não recebe aumento.
df_silver = df_bronze.filter(col("salario").isNotNull()) \
.filter(col("departamento") != "Gestão") \
.withColumn("salario_ajustado", col("salario") * 1.10)

df_silver.createOrReplaceTempView("v_funcionarios_silver")
print("Dados transformados no PySpark!")
```

```python
df_silver.display()
```

```python
%sql
SELECT
departamento,
count(*) as total_funcionarios,
avg(salario_ajustado) as media_salarial
FROM v_funcionarios_silver
GROUP BY departamento
```

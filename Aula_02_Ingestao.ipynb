**Usei essa tabela com dataset:** https://www.kaggle.com/datasets/vinothkannaece/sales-dataset/data

```python
from pyspark.sql.functions import current_timestamp, from_utc_timestamp

tblVendas = "/Volumes/workspace/default/aula_data_engineer/sales_data.csv" 

df_vendas = spark.read \
    .format("csv") \
    .option("header", "true") \
    .load(tblVendas)

df_transformado = df_vendas.withColumn("data_processamento", from_utc_timestamp(current_timestamp(), "America/Sao_Paulo"))


df_transformado.write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/aula_data_engineer/sales_silver")

display(df_transformado)
```
